
# Luego importamos el Data Set de las 50 Startups


```python
dataset = pd.read_csv('50_Startups.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 4].values
```

# Seguidamente codificamos las columnas categóricas, en este caso solo tenemos una: Estado que es la 4ta columna, es decir la indexada con 3. Es importante resaltar que primero con el LabelEncoder la codificamos es decir la transformamos a números, pero como ya explicamos en este caso no hay orden entre ciudades, asi es que debemos implementar las variables Dummies, para eso usamos el OneHotEncoder

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import make_column_transformer
labelencoder_X = LabelEncoder()
X[:, 3] = labelencoder_X.fit_transform(X[:, 3])
onehotencoder = make_column_transformer((OneHotEncoder(), [3]), remainder = "passthrough")
X = onehotencoder.fit_transform(X)
```

# Ahora nos toca evitar la trampa de las variables Dummies (ficticias) y de esta forma evitar la colinealidad. Para eso decidimos eliminar la primera variable Dummy correspondiente a California, entonces estructuramos nuestro DataFrame a partir la columna 1 solamente, ignorando la 0.

```python
X = X[:, 1:]
```
